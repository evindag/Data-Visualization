#Workshop

```python
x = [1,2,3,4,5,6,7,8]
y = [1,4,9,16,25,36,49,64]

x2 = [1,2,3,4,5,6,7,8]
y2 = [1,8,27,64,125,216,343,512]
```

```python
plt.plot(x, y , label= "X Kare", color="red")
plt.plot(x2, y2, label= "X Küp")
plt.title("Line Plot")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.legend()
plt.grid()
plt.show()
```

#Subplots

```python
fig = plt.figure(figsize=(18,5))
first_plot = fig.add_subplot(1,3,1)
first_plot.plot(x,y, color="red")
first_plot.set_title("Line Plot")

second_plot = fig.add_subplot(1,3,2)
second_plot.scatter(x,y)

third_plot = fig.add_subplot(1,3,3)
third_plot.bar(x,y)

plt.show()
```
