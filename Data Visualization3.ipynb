```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
data = diamonds.copy()
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
data['cut'].value_counts()
```

```python
data['color'].value_counts()
```

```python
from pandas.api.types import CategoricalDtype
```

```python
data.cut.head()
```

```python
data.cut = data.cut.astype(CategoricalDtype(ordered=True))
```

```python
data.cut.head()
```

```python
data.cut.value_counts().plot.bar()
```

```python
data.cut.value_counts().plot.barh().set_title("Cut Column")
```

```python
data.cut.value_counts().plot.barh().set_title("Cut Column");
```

```python
cut_category = ["Fair","Good","Very Good", "Premium", "Ideal"]
```

```python
data.cut = data.cut.astype(CategoricalDtype(categories=cut_category,ordered=True))
```

```python
data.cut.head()
```

```python
data["cut"].value_counts().plot.barh().set_title("Cut Column");
```

```python
sns.barplot(x='cut',y=data.cut.index, data=data)
```

```python
sns.catplot(x='cut',y='price',data=data)
```

```python
sns.catplot(x='cut',y='price',hue='color',data=data)
```

```python
sns.barplot(x='cut',y='price',hue='color',data=data)
```
